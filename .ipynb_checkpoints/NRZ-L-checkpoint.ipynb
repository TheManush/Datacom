```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits_str = '10111001'
bits_int = np.array([int(char) for char in bits_str])
print("Binary sequence:", bits_int)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Original Signal", fontdict={'fontsize': 20})  
plt.step(range(len(bits_int)+1), np.append(bits_int, bits_int[-1]), where='post', 
         color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})  
plt.ylabel('Voltage', fontdict={'fontsize': 15})  
plt.grid(True)
plt.show()
```

```python
encoded = []
for bit in bits_int:
    if bit == 1:
        encoded.append(1)  # High for 1
    else:
        encoded.append(0)  # Low for 0

nrz_l = np.array(encoded)  # Convert to numpy array if needed
print("NRZ-L encoded:", nrz_l)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-L Encoded Signal", fontdict={'fontsize': 20})
plt.step(range(len(nrz_l)+1), np.append(nrz_l, nrz_l[-1]), where='post',
         color='red', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontsize': 15})
plt.grid(True)
plt.show()
```

```python
#Decode
decoded = []
for bit in bits_int:
    if bit == 1:
        decoded.append(1)  # High for 1
    else:
        decoded.append(0)  # Low for 0

nrz_l2 = np.array(decoded)  
print("NRZ-L decoded:", nrz_l2)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-L Decoded Signal", fontdict={'fontsize': 20})
plt.step(range(len(nrz_l2)+1), np.append(nrz_l2, nrz_l2[-1]), where='post',
         color='green', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontsize': 15})
plt.grid(True)
plt.show()
```
