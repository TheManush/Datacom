```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits_str = '10111001'
bits_int = np.array([int(char) for char in bits_str])
print("Binary sequence:", bits_int)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Og Signal", fontdict={'fontsize': 20})  
plt.step(range(len(bits_int)+1), np.append(bits_int, bits_int[-1]), where='post', 
         color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})  
plt.ylabel('Voltage', fontdict={'fontsize': 15})  
plt.grid(True)
plt.show()
```

```python
nrz_i = np.zeros_like(bits_int)
current_level = 1 #starts at 1  

for i, bit in enumerate(bits_int):
    if bit == 1:
        current_level = 1 - current_level  # Invert on '1'
    # else keep the same level
    nrz_i[i] = current_level

print("NRZ-I encoded:", nrz_i)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-I Encoded Signal", fontdict={'fontsize': 20})

# Step plot
plt.step(range(len(nrz_i)+1), np.append(nrz_i, nrz_i[-1]), where='post',
         color='red', linewidth=2)

plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontsize': 15})
plt.grid(True)
plt.show()
```

```python
decoded_bits = []
prev_level = 1 

for level in nrz_i:
    if level != prev_level:
        decoded_bits.append(1)
    else:
        decoded_bits.append(0)
    prev_level = level 
decoded_array = np.array(decoded_bits)
print("NRZ-I decoded:", decoded_array)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-I Decoded Signal", fontdict={'fontsize': 20})

plt.step(range(len(decoded_array)+1), np.append(decoded_array,decoded_array[-1]), where='post',
         color='green', linewidth=2)

plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontsize': 15})
plt.grid(True)
plt.show()
```
