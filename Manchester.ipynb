```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bits_str = '10111001'
bits_int = np.array([int(char) for char in bits_str])
print("Binary sequence:", bits_int)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Original Signal", fontdict={'fontsize': 20})  
plt.step(range(len(bits_int)+1), np.append(bits_int, bits_int[-1]), where='post', 
         color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})  
plt.ylabel('Voltage', fontdict={'fontsize': 15})  
plt.grid(True)
plt.show()
```

```python
encoded = []
for bit in bits_int:
    if bit == 1:
        encoded.append(1)
        encoded.append(0)
    else:
        encoded.append(0) 
        encoded.append(1)

manc = np.array(encoded) 
print("Manchester encoded:", manc)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-L Encoded Signal", fontdict={'fontsize': 20})
plt.step(range(len(manc)+1), np.append(manc, manc[-1]), where='post',
         color='red', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontsize': 15})
plt.grid(True)
plt.show()
```

```python
decoded= []  
for i in range(0, len(manc), 2):
    if i + 1 < len(manc):
        curr = manc[i]
        next_v = manc[i + 1]
        
        if curr== 1 and next_v == 0:
            decoded.append(1)
        elif curr == 0 and next_v == 1:
            decoded.append(0)

manc2 = np.array(decoded)
print("Manchester decoded:", manc2)
```

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-L Encoded Signal", fontdict={'fontsize': 20})
plt.step(range(len(manc2)+1), np.append(manc2, manc2[-1]), where='post',
         color='green', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontsize': 15})
plt.grid(True)
plt.show()
```
